```python
## Perform a hypothesis test and state the confidence level at which you can confirm a correlation between the two variables (use the scipy.stats.pearsonr() function--consult the SciPy documentation for details (Links to an external site.)Links to an external site. about its inputs and outputs).
```

```python
from scipy import stats
```

```python
stats.pearsonr(tf.Gestational_Age,tf.Birth_Weight)
```

```python
## H0:Gestational_Age & Birth_Weight are not correlated (pear corr=0)
## H1:Gestational_Age & Birth_Weight are correlated ((pear corr> or <0))
## Above pearson correlation values indicate p<alpha(5.6210785283336966e-05) and correlation is 0.8. Hence reject null hypothesis
## H0 with confidence level of 95%.
```
